Heart Disease Prediction

```python
import pandas as pd
import numpy as np
url="E:\heart disease dataset.zip"
data=pd.read_csv(url,compression='zip')
df=pd.DataFrame(data)
df.head()
df.describe()
df.isnull().sum()
x=df.iloc[:,:-1]
y=df["target"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)

from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.svm import SVC
model=SVC(kernel='linear')
model.fit(x_train,y_train)
y_pred_linear=model.predict(x_test)
print("Linear =accuracy score:",accuracy_score(y_test,y_pred_linear))
print("Linear =classification record:",classification_report(y_test,y_pred_linear))
print("Linear=confusion matric:",confusion_matrix(y_test,y_pred_linear))

#polynominal
model=SVC(kernel='poly')
model.fit(x_train,y_train)
y_pred_poly=model.predict(x_test)
print("polynominal =accuracy score:",accuracy_score(y_test,y_pred_poly))
print("Polynominal=classification record:",classification_report(y_test,y_pred_poly))
print("polynominal=confusion matric:",confusion_matrix(y_test,y_pred_poly))

#rbf
model=SVC(kernel='rbf')
model.fit(x_train,y_train)
y_pred_rbf=model.predict(x_test)
print("RBF =accuracy score:",accuracy_score(y_test,y_pred_rbf))
print("RBF=classification record:",classification_report(y_test,y_pred_rbf))
print("RBF=confusion matric:",confusion_matrix(y_test,y_pred_rbf))

#Sigmoid
model=SVC(kernel="sigmoid")
model.fit(x_train,y_train)
y_pred_sigmoid=model.predict(x_test)
print("Sigmoid =accuracy score:",accuracy_score(y_test,y_pred_sigmoid))
print("Sigmoid=classification record:",classification_report(y_test,y_pred_sigmoid))
print("Sigmoid=confusion matric:",confusion_matrix(y_test,y_pred_sigmoid))

import matplotlib.pyplot as plt

kernels = ['Linear', 'Polynomial', 'RBF', 'Sigmoid']
accuracy = [81.46, 90.73, 88.78, 74.63]

plt.figure(figsize=(8,5))
plt.bar(kernels, accuracy)

plt.xlabel('Kernel')
plt.ylabel('Accuracy (%)')
plt.title('SVM Kernel Comparison')

for i,v in enumerate(accuracy):
    plt.text(i, v+0.5, str(v))

plt.show()
```
